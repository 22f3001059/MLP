```python
#question 11
# i will be using apply method and a function to make a new column

def get_category(x):
    if x < 10:
        return "Category 1"
    elif 10 <= x < 20:
        return "Category 2"
    elif 20 <= x < 30:
        return "Category 3"
    elif 30 <= x < 40:
        return "Category 4"
    return "Category 5"

df["CATEGORY"] = df["PRICE"].apply(get_category)
df["CATEGORY"].value_counts()
```

```python
import numpy as np


df["RM"] = df["RM"].replace(-1, np.nan)
df["AGE"] = df["AGE"].replace(-2, np.nan)
df["RIVERSIDE"] = df["RIVERSIDE"].replace("UNKNOWN", np.nan)
#now we can just get rid of the unknowns to get our datasets that we will work on
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns="PRICE")
y = df["PRICE"]

X.head()
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train.shape, X_test.shape
```

```python
#question 13: loaded fking question, i'll use pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder


X_train.drop(columns="CATEGORY", inplace=True)
X_test = X_test.drop(columns="CATEGORY")

X_train.head()
```

```python
#now we create a pipeline for transformations

transformer = ColumnTransformer([

    ("mnscaler", MinMaxScaler(), ["CRIM", "ZN","POLINDEX","DIS","HIGHWAYCOUNT","TAX","PTRATIO","IMM","BPL"]),
    ("indus_stdscaler", StandardScaler(), ["INDUS"]),
    ("RM_pipeline", Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", MinMaxScaler())
    ]), ["RM"]) ,
    ("AGE_pipeline", Pipeline([
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", MinMaxScaler())
    ]), ["AGE"]),
    ("riverside_pipeline", Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder())
    ]), ["RIVERSIDE"]),
],
                                remainder="passthrough"
)

transformer
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
X_train.shape
```
